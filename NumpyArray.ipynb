```python
array2 = np.array([[['a','b'],['c','d']],
                   [['e','f'],['g','h']],                  #This is a 3D Array
                   [['i','j'],['k','l']]])
```

```python
array2.shape        # Method to identify array details
```

```python
array2.ndim     # method to identifies array dimention
```

```python
array2[0][1][0]     #print array value
```

```python
array2[0,1,0] #we dont need different parenthesis for print
```

```python
word = array2[0,0,0] + array2[2,1,1] + array2[2,1,1]
word
```
